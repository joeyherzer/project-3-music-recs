```python
import pandas as pd
```

```python
listens = pd.read_csv('user_artists.dat', sep='\t')
listens.head()
```

```python
artistList = pd.read_csv('artists.dat', sep='\t')

artistList= artistList.drop(columns=['url','pictureURL'])
artistList.head()
```

```python
listens = listens.merge(artistList, left_on='artistID', right_on='id')
listens = listens.drop(columns='id')
listens = listens.rename(columns = 
                         {'name':'artist',
                          'weight':'user_listens'
                         })
listens.head()
```

```python
userTotals = listens.groupby(['userID']).sum()
userTotals = userTotals.drop(columns='artistID')
userTotals = userTotals.rename(columns = {'user_listens':'total_listens'})
userTotals.head()
```

```python
normalListens = listens.merge(userTotals, left_on = 'userID', right_on='userID')
normalListens
```

```python
normalListens['weighted_listens'] = normalListens['user_listens'] / normalListens['total_listens']
normalListens
```
